```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate linespace point
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x-0),'--',color='black')   
plt.plot(x,np.cos(x-1),'+',color='green')   
plt.plot(x,np.cos(x-2),'*',color='0.89')     
plt.plot(x,np.sin(x-3),'-',color='#ABCDEF')
plt.show
```

```python
x=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x-0),'--',color='black')
plt.plot(x,np.cos(x-1),'+',color='green')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#generate some data
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)

#create subplot
fig,(ax1,ax2)=plt.subplots(2)

#plot data on the first subplot
ax1.plot(x,y1)
ax2.plot(x,y2)
ax1.set_title('sin(x)')
ax2.set_title('cos(x)')
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
#generate some data
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)

#create subplot
fig,(ax1,ax2)=plt.subplots(2)

#plot data on the first subplot
ax1.plot(x,y1)
ax2.plot(x,y2)
ax1.set_title('sin(x)')
ax2.set_title('cos(x)')
```
